```python
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
#9
sns.catplot(x ='class', hue='sex', kind='count', data = titanic) 
```

```python
#more surviving classes
#1
sns.catplot(x ='class', hue='survived', kind='count', data = titanic) 
```

```python

#2
sns.barplot(x='sex', y='fare', data=titanic)

```

```python
#3
sns.histplot(titanic['age'], bins=10)
plot.title('age ')
plot.show()
```

```python
#4
plot.scatter(titanic['age'], titanic['fare'])
plot.xlabel('age')
plot.ylabel('fare')
plot.title('age vs fare')
plot.show()
```

```python
#5
sns.catplot(x ='embark_town', hue='survived', kind='count', data = titanic) 
```

```python
#6
sns.barplot(x ="sex", y ="age", hue ="survived",  data = titanic) 
```

```python
#7 s
sns.catplot(x ="who", hue ="survived",  
kind ="count", data = titanic) 
```

```python
#8
#Male or Female
who = titanic['who'].value_counts()
plot.pie(who, labels = who.index)
```
