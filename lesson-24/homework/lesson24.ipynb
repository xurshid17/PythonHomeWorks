```python
import pandas as pd 
import numpy as np
```

```python
sales_data = pd.read_csv('task\\sales_data.csv')
sales_data
```

```python
#1
category = (sales_data
                .groupby('Category')
                .agg(total_sol = pd.NamedAgg(column='Quantity', aggfunc='sum'),
                     average_price = pd.NamedAgg(column='Price', aggfunc='mean'),
                     max_quantity_sold = pd.NamedAgg(column='Quantity', aggfunc='max'))
                     .reset_index())
category
```

```python
#2
sell = (sales_data
                .groupby(['Category', 'Product'])
                .agg(total_sol = pd.NamedAgg(column='Quantity', aggfunc='sum'))
                     .reset_index())

top_sell =(sell
           .groupby(['Category', 'Product'])
           .agg({'total_sol':'idxmax'}))
top_sell
```

```python
#3
sales_data['total_sale'] = sales_data['Quantity'] * sales_data['Price']
date_sales = sales_data.groupby('Date')['total_sale'].sum().reset_index()

max_sales_date = date_sales.loc[date_sales['total_sale'].idxmax()]
max_sales_date
```

```python
customer_orders = pd.read_csv('task\\customer_orders.csv')
customer_orders
```

```python
category_info = (customer_orders
                .groupby('CustomerID')
                .agg({'OrderID':'count'})
                      .astype(int))
fil_cus = category_info[category_info['OrderID'] >= 20]
fil_cus        
```

```python
#2
ave_price = (customer_orders
             .groupby('CustomerID')
             .agg({'Price':'mean'}))
ave_price[ave_price['Price'] > 120] 
```

```python
#3
quantity_price = (customer_orders
                  .groupby('Product')
                  .agg({'Quantity': 'sum',
                       'Price':'sum'}))
quantity_price[quantity_price['Quantity'] >= 5]
```
